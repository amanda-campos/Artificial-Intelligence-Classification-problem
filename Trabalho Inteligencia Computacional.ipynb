```python
#MATRIZ DE DISTÃ‚NCIAS
import numpy as np 
import pandas as pd
import seaborn as sns
import scipy.spatial as spatial
df = pd.read_csv(r"C:\Users\Amanda\Desktop\bank-additional.csv",sep=';')
df['y'] = LabelEncoder().fit_transform(df['y'])
df.head()
numeric_df = df.select_dtypes(exclude="object")

pdistmat = spatial.distance.squareform(spatial.distance.pdist(numeric_df))

N = len(numeric_df)
plt.pcolormesh(pdistmat)
plt.colorbar()
plt.xlim([0,N])
plt.ylim([0,N])
plt.show()
print(N)
```
